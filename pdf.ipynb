```python
from elasticsearch import Elasticsearch
from elasticsearch_dsl import Document, Text
from elasticsearch_dsl.connections import connections
from PyPDF2 import PdfReader

# 建立到Elasticsearch的连接
connections.create_connection(hosts=['http://localhost:9200'])

# 定义Elasticsearch索引
class PDFDocument(Document):
    title = Text()
    content = Text()

    class Index:
        name = 'pdf_index'

# 创建索引
PDFDocument.init()

# 读取PDF文件并索引内容
def index_pdf(pdf_path):
    with open(pdf_path, 'rb') as file:
        pdf = PdfReader(file)
        text = ''
        for page_num in range(len(pdf.pages)):
            page = pdf.pages[page_num]
            text += page.extract_text()

    doc = PDFDocument(title=pdf_path, content=text)
    doc.save()

# 将PDF文件索引到Elasticsearch中
index_pdf('data/子平真诠.pdf')
```

```python
from elasticsearch_dsl import Search
from PyPDF2 import PdfReader

def search_pdf(keyword_):
    # 创建搜索对象
    s = Search(index='pdf_index')

    # 执行搜索
    response = s.query('match', content=keyword_).execute()

    # 显示搜索结果
    for hit in response:
        print("Title:", hit.title)
        print("Content:")
        paragraphs = hit.content.split("\n")  # 假设每个段落之间有两个换行符分隔
        for paragraph in paragraphs:
            sentences = paragraph.split(". ")  # 假设句子之间以句号和空格分隔
            for sentence in sentences:
                if keyword_ in sentence:
                    print(sentence)
                    print()
                    break  # 只显示第一个包含关键词的句子

search_pdf('壬')
```
